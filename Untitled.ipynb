```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Data/YouGov_2024_general_election_MRP_2.xlsx'


df = pd.read_excel(file_path)
df.head(1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
duplicates = df.duplicated()
print(f'Duplicates: {duplicates.sum()}')
```

```python
#This cell is checking if the sum of the %(percentages) is 100%, LOOKING for errors. 
percentage_columns = ['ConShare', 'LabShare', 'LibDemShare', 'GreenShare', 'ReformShare', 'PlaidShare', 'SNPShare', 'OthersShare']
from IPython.display import display

df['TotalShare'] = df[percentage_columns].sum(axis=1)


invalid_rows = df[df['TotalShare'] != 1]

# print(invalid_rows)

```

```python
#Renaming columns
df.columns = [
    'Area', 'Region', 'Const', 'ConShare', 'LabShare', 'LibDemShare', 'GreenShare', 
    'ReformShare', 'PlaidShare', 'SNPShare', 'OthersShare', 'Winner2024', 'Second2024', 
    'YouGovCall', 'WinnerMargin', 'Status2019v2024', 'Win2019Nominal', 'ConstituencyType', 'TotalShare'
]

df.head()
```

```python
# Checking const for duplicates since the Const is serving as unique id
check_const = df[df['Const'].duplicated(keep=False)]
check_const
```

```python
plt.figure(figsize=(7, 4))
winner_counts = df['Winner2024'].value_counts()
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette='viridis')
plt.title('Number of Constituencies Won by Each Party')
plt.xlabel('Party')
plt.ylabel('Number of Constituencies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for overall vote share
total_shares = df[['ConShare', 'LabShare', 'LibDemShare', 'GreenShare', 'ReformShare', 'PlaidShare', 'SNPShare', 'OthersShare']].sum()
labels = total_shares.index
sizes = total_shares.values
plt.figure(figsize=(7, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(labels)))
plt.title('Overall Vote Share by Each Party')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Box plot for cheking the distribution of vote shares
plt.figure(figsize=(7, 4))
sns.boxplot(data=df[['ConShare', 'LabShare', 'LibDemShare', 'GreenShare', 'ReformShare', 'PlaidShare', 'SNPShare', 'OthersShare']])
plt.title('Distribution of Vote Shares by Party')
plt.ylabel('Vote Share')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of vote shares by region to check any possible CORRELATION
region_shares = df.groupby('Region')[['ConShare', 'LabShare', 'LibDemShare', 'GreenShare', 'ReformShare', 'PlaidShare', 'SNPShare', 'OthersShare']].mean()

plt.figure(figsize=(8, 5))
sns.heatmap(region_shares, annot=True, cmap='viridis', linewidths=.5)
plt.title('Average Vote Shares by Region')
plt.xlabel('Party')
plt.ylabel('Region')
plt.xticks(rotation=45)
plt.show()
```

```python
#Saving the cleaned df into excel file 
output_file_path = 'Data/YouGov_2024_general_election_MRP_2_cleaned.xlsx'
df.to_excel(output_file_path, index=False)
```
